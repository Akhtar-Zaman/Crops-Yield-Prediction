#Area vs Production


```python
dataset = pd.read_csv(r'/content/drive/My Drive/Python Scripts/Dataset/Production-Area.csv')


plt.figure(figsize=(7,4))
plt.scatter(dataset.Aman_Area, dataset.Aman_Production, label='Aman', linewidth=2.0)
plt.scatter(dataset.Aus_Area, dataset.Aus_Production, label='Aus', linewidth=2.0)
plt.scatter(dataset.Boro_Area, dataset.Boro_Production, label='Boro', linewidth=2.0)
plt.scatter(dataset.Jute_Area, dataset.Jute_Production, label='Jute', linewidth=2.0)
plt.scatter(dataset.Potato_Area, dataset.Potato_Production, label='Potato', linewidth=2.0)
plt.scatter(dataset.Wheat_Area, dataset.Wheat_Production, label='Wheat', linewidth=2.0)

#plt.title("Aus", fontsize=20)
plt.yscale('symlog')
plt.xscale('symlog')
plt.xlabel("Area (acres)", fontsize=18)
#plt.xticks([100000,200000,300000,400000,500000,600000,700000])
plt.ylabel("Crop Yield (M.tons)", fontsize=18)
#plt.ylim(25,40)
plt.legend(loc='lower right', ncol=3)
plt.savefig("Annual2.pdf")
files.download("Annual2.pdf")
plt.show()



 
```

#Temperature

```python
dataset = pd.read_csv(r'/content/drive/My Drive/Python Scripts/Dataset/Temp.csv')
plt.figure(figsize=(7,4))
plt.plot(dataset.YEAR, dataset.Aman, label='Aman', linewidth=4.0)
plt.plot(dataset.YEAR, dataset.Aus, label='Aus', linewidth=4.0)
plt.plot(dataset.YEAR, dataset.Boro, label='Boro', linewidth=4.0)
plt.plot(dataset.YEAR, dataset.Jute, label='Jute', linewidth=4.0)
plt.plot(dataset.YEAR, dataset.Potato, label='Potato', linewidth=4.0)
plt.plot(dataset.YEAR, dataset.Wheat, label='Wheat', linewidth=4.0)

#plt.title("Aus", fontsize=20)
#plt.yscale('symlog')
plt.xlabel("Year", fontsize=18)
plt.xticks([1969,1974,1979,1984,1989,1994,1999,2004,2009,2013])
plt.ylabel("Max-Temp (°C)", fontsize=18)
plt.ylim(25,40)
plt.legend(loc='lower right', ncol=3)
plt.savefig("Temp.pdf")
files.download("Temp.pdf")
plt.show()



 
```
